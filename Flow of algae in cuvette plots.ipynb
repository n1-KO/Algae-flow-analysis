## Plot vectors

Make a plot with vectors indicating the direction of low for each frame. Do a vectorial sum to obtain a vector indicating the direction of flow of the algae for each light intensity.


### Raw data

```python
#the v_sum is plotted in a separate figure if modules not inserted in this cell
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

#function for the plot
def vector_plot(filename, x0, x1, title,savename):
    '''Function for plotting the vectors
    Input:
    filename: name of file from which data is extracted
    x0: first frame
    x1: last frame
    '''

    #file from PIVLab (matlab) indicating the mean flow over the selected area of the cuvette
    frame,magnitude,angle= np.loadtxt(filename, delimiter=',', unpack=True)

    #extract y and x values for each vector
    x = np.cos(angle)*magnitude
    y = np.sin(angle)*magnitude
    origin = [0], [0] # origin point

    #plot it:
    plt.figure() #generate new figure for each intensity
    plt.quiver(*origin, x[x0:x1], y[x0:x1]) #vectors for the frames selected
    plt.xlabel('xaxis')
    plt.ylabel('yaxis')
    plt.title(title)
    plt.axvline(x = 0, c='b', ls='-',lw=1,alpha=0.25, zorder=0)
    plt.axhline(y = 0, c='b', ls='-',lw=1,alpha=0.25, zorder=0)
    plt.show()
    #vector resulting from adding all the vectors together (for selected frames):       
    v_sum = np.sum(x[x0:x1]),np.sum(y[x0:x1])
    print('The vectorial sum is the vector:', v_sum)
    print('The magnitude of the vectorial sum vector is:', np.linalg.norm(v_sum))
    plt.quiver(*origin, v_sum[0], v_sum[1], color=['r'], zorder=1)
    #print the direction of swim. Bear in mind that the video with the cuvette is tilted 90 degrees 
    #to the left. Thus vectors pointing in positive x direction indicate flow downwards, whereas
    #vectors pointing in negative x direction indicate flow upwards
    if v_sum[0]>0:
        print('The algae are swimming away from the light (negative photoaxis)')
    else:
        print('The algae are swimming towards the light (positive photoaxis)')
    #save the plot
    plt.savefig(savename)
    #need to omit magnitudes exceeding cerain value (done more precisely later):
    #print('maximum magnitude is:', max(magnitude))
```

```python
#plot it for the different intensities (1st experiment):
vector_plot('PIVlab_Extr_mean_flow_2previous.txt', 1,150, 'intensity2', 'intensity2.png')
vector_plot('PIVlab_Extr_mean_flow_2previous.txt', 151,300, 'intensity3', 'intensity3.png')
vector_plot('PIVlab_Extr_mean_flow_2previous.txt', 301,450, 'intensity4', 'intensity4.png')
vector_plot('PIVlab_Extr_mean_flow_2previous.txt', 451,599, 'intensity5','intensity5.png') 
#last frame corresponds to a disproportional vector in magnitude(error?)

#data from second experiment:
vector_plot('PIVlab_Extr_mean_flow_video1.txt', 1, 299, 'Power: 0,45 mW \n Irradiance: 365 Î¼W*cm^2', 'Power045.png')
vector_plot('PIVlab_Extr_mean_flow_2.txt', 1, 299, 'Power: 0,92 mW \n Irradiance: 720 Î¼W*cm^2', 'Power092.png')
vector_plot('PIVlab_Extr_mean_flow_34.txt', 1, 299, 'Power: 1,39 mW \n Irradiance: 1,09 mW*cm^2', 'Power139.png')
vector_plot('PIVlab_Extr_mean_flow_34.txt', 300, 599, 'Power: 1,75 mW \n Irradiance: 1,37 mW*cm^2', 'Power175.png')
vector_plot('PIVlab_Extr_mean_flow_57.txt', 1, 299, 'Power: 2,0 mW \n Irradiance: 1,59 mW*cm^2', 'Power20.png')
vector_plot('PIVlab_Extr_mean_flow_57.txt', 300, 599, 'Power: 2,5 mW \n Irradiance: 1,98 mW*cm^2', 'Power25.png')
vector_plot('PIVlab_Extr_mean_flow_8.txt', 1, 299, 'Power: 3,0 mW \n Irradiance: 2,37 mW*cm^2', 'Power30.png')
vector_plot('PIVlab_Extr_mean_flow_9.txt', 1, 299, 'Power: 3,5 mW \n Irradiance: 2,73 mW*cm^2', 'Power35.png')
vector_plot('PIVlab_Extr_mean_flow_11.txt', 1, 299, 'Power: 4,0 mW \n Irradiance: 3,15 mW*cm^2', 'Power40.png')
vector_plot('PIVlab_Extr_mean_flow_12.txt', 1, 299, 'Power: 4,5 mW \n Irradiance: 3,53 mW*cm^2', 'Power45.png')
vector_plot('PIVlab_Extr_mean_flow_13.txt', 1, 299, 'Power: 5,0 mW \n Irradiance: 3,95 mW*cm^2', 'Power50.png')
```

### Corrected data

As seen above the data from the second experiment seems to not match very well the predictions made based on the 1st run. Moreover, there seems to be a several vectors with greater magnitude from the rest. In order to correct the data, and eliminate these vectors, a gaussian distribution of the magnitudes was analyzed and a code was done to only consider the data within a determined value of standard deviation. 


#### Draw the histograms 

```python
#first draw historgams to see distribution of the data
def draw_hist(file, ini_fram, fin_fram, title1, hist_save_name):
    '''Function that plots the histograms of vector magnitude corresponding to each data file
    '''
    #file from PIVLab (matlab) indicating the mean flow over the selected area of the cuvette
    frame,magnitude,angle= np.loadtxt(file, delimiter=',', unpack=True)
    print('The mean of the data is', np.mean(magnitude [ini_fram:fin_fram]))
    
    
    #plot the histogram
    plt.figure()
    plt.hist(magnitude [ini_fram:fin_fram],bins=50,density= True, edgecolor='k') #density=True normalizes the bins
    plt.title(title1)
    plt.xlabel("value")
    plt.ylabel("number of occurences")
    
    #import the module for the gaussian
    import scipy.stats as stats
    x_0, sigma = stats.norm.fit(magnitude[ini_fram:fin_fram])
    print ("Fitted Gaussian: \n Mean value ", x_0, "with standard deviation", sigma)
    x = np.linspace (-0.0001,0.0002,100) #array of x values used for plotting the gausian distribution curve
    gaussian_curve = stats.norm.pdf(x,x_0,sigma) #the y values
    plt.plot(x,gaussian_curve,'r-.', label="stats.norm.pdf")
    #plt.axvline(x = (x_0+sigma), c='k', ls='-',lw=1,alpha=1, zorder=0)
    #plt.axvline(x = (x_0-sigma), c='k', ls='-',lw=1,alpha=1, zorder=0)
    #save the plot
    plt.savefig(hist_save_name) 
```

#### Draw the vectors

A new code needed to be written for this part. This one loaded the data separating it into rows as desired for this part (need to exclude certain magnitudes or data outliers).

```python
def data_correction(file_name, ini_frame, fin_frame, title_1, save_name):
    '''New function to load the data and calculate the mean value and standar deviation
    '''
    #open the file
    with open(file_name) as fp: data = fp.read().splitlines()
    #print(data)
    
    #create an array for very data point: [frame nomber, vector magnitude, vector angle]
    DATA =[]
    for dat in data:
        dat_split = dat.split(',')
        dat_new = []
        for d in dat_split:
            dat_new.append(float(d))
        DATA.append(dat_new)

    #print(DATA)

    #extract the magnitude of the original data ('DATA')
    magnitude1=[]
    for i in range(len(DATA)):
        magnitude1.append(DATA[i][1])

    #calculate the mean value and standard deviation of the magnitude of the original data
    import scipy.stats as stats
    x_0, sigma = stats.norm.fit(magnitude1)
    print ("Fitted Gaussian: \n Mean value ", x_0, "with standard deviation", sigma)

    #discard data points outside of the value of standard deviatio desired:
    def data_elim (sigma_no):
        '''Function made to select points of data within x standard deviations from the mean.
        Inputs:
        -sigma_no: number of sigma values that determines the range of the new data
        '''
        #loop for the selection of data
        #initialise with an empty array
        data_new =[]
        for i in range (len(DATA)):
             if DATA[i][1] <= (x_0+(sigma_no*sigma)): 
                    if DATA[i][1] >= (x_0-(sigma_no*sigma)):
                        data_new.append(DATA[i])
                        
        print ('The data is in range of',sigma_no, 'standard deviations')
                        
        #magnitude values of new data ('data_new')              
        magnitude3=[]
        for i in range(len(data_new)):
            magnitude3.append(data_new[i][1])
        #print(magnitude3,'\n')

        #angle values of new data ('data_new')
        angle=[]
        for i in range(len(data_new)):
            angle.append(data_new[i][2])
        #print(angle)

        #x and y values of the new data set
        xvalues =[]
        yvalues =[]
        for i in range (len(data_new)):
            xvalues.append(np.cos(data_new[i][2])* data_new[i][1])
            yvalues.append(np.sin(data_new[i][2])* data_new[i][1])
            
        origin = [0], [0] #origin point

        #plot it:
        plt.figure() #generate new figure for each intensity
        plt.quiver(*origin, xvalues[ini_frame:fin_frame], yvalues[ini_frame:fin_frame]) #vectors for the frames selected
        plt.xlabel('xaxis')
        plt.ylabel('yaxis')
        plt.title(title_1)
        plt.axvline(x = 0, c='b', ls='-',lw=1,alpha=0.25, zorder=0)
        plt.axhline(y = 0, c='b', ls='-',lw=1,alpha=0.25, zorder=0)
        plt.show()
        #vector resulting from adding all the vectors together (for selected frames):       
        v_sum = np.sum(xvalues[ini_frame:fin_frame]),np.sum(yvalues[ini_frame:fin_frame])
        #print('The vectorial sum is the vector:', v_sum)
        #print('The magnitude of the vectorial sum vector is:', np.linalg.norm(v_sum))
        plt.quiver(*origin, v_sum[0], v_sum[1], color=['r'], zorder=1)
        #print the direction of swim. Bear in mind that the video with the cuvette is tilted 90 degrees 
        #to the left. Thus vectors pointing in positive x direction indicate flow downwards, whereas
        #vectors pointing in negative x direction indicate flow upwards
        if v_sum[0]>0:
            print('The algae are swimming away from the light (negative photoaxis)')
        else:
            print('The algae are swimming towards the light (positive photoaxis)')
        #determine the confidence interval of the data:
        if sigma_no==1:
            print ('The confidence interval is of 68%')
        if sigma_no==2:
            print ('The confidence interval is of 95%')
        if sigma_no==3:
            print ('The confidence interval is of 99.7%')
        if sigma_no>3:
            print ('The confidence interval is of 100%')
        print()
              
        #save the plot
        plt.savefig(save_name) 
    #set sigma multiplicity for the new data
    data_elim(2)
    
    
```

```python
#data from the second experiment:
data_correction('PIVlab_Extr_mean_flow_video1.txt', 1, 299, 'Power: 0,45 mW \n Irradiance: 365 Î¼W*cm^2', 'c_power0045.png')
data_correction('PIVlab_Extr_mean_flow_2.txt', 1, 299, 'Power: 0,92 mW \n Irradiance: 720 Î¼W*cm^2', 'c_power092.png')
data_correction('PIVlab_Extr_mean_flow_34.txt', 1, 299, 'Power: 1,39 mW \n Irradiance: 1,09 mW*cm^2', 'c_power139.png')
data_correction('PIVlab_Extr_mean_flow_34.txt', 300, 599, 'Power: 1,75 mW \n Irradiance: 1,37 mW*cm^2', 'c_power175.png')
data_correction('PIVlab_Extr_mean_flow_57.txt', 1, 299, 'Power: 2,0 mW \n Irradiance: 1,59 mW*cm^2', 'c_power20.png')
data_correction('PIVlab_Extr_mean_flow_57.txt', 300, 599, 'Power: 2,5 mW \n Irradiance: 1,98 mW*cm^2', 'c_power25.png')
data_correction('PIVlab_Extr_mean_flow_8.txt', 1, 299, 'Power: 3,0 mW \n Irradiance: 2,37 mW*cm^2', 'c_power30.png')
data_correction('PIVlab_Extr_mean_flow_9.txt', 1, 299, 'Power: 3,5 mW \n Irradiance: 2,73 mW*cm^2', 'c_power35.png')
data_correction('PIVlab_Extr_mean_flow_11.txt', 1, 299, 'Power: 4,0 mW \n Irradiance: 3,15 mW*cm^2', 'c_power40.png')
data_correction('PIVlab_Extr_mean_flow_12.txt', 1, 299, 'Power: 4,5 mW \n Irradiance: 3,53 mW*cm^2', 'c_power45.png')
data_correction('PIVlab_Extr_mean_flow_13.txt', 1, 299, 'Power: 5,0 mW \n Irradiance: 3,95 mW*cm^2', 'c_power50.png')
```
